```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set(color_codes=True)
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from sklearn.preprocessing import *
```

```python
#Generating normally distributed data

df = pd.DataFrame({
    'x1':np.random.normal(0, 2, 10000),
    'x2':np.random.normal(5, 3, 10000),
    'x3':np.random.normal(-5, 5, 10000)})
```

```python
df.plot.kde() # 열 단위로 선 그래프를 그리기
```

```python
data_tf = StandardScaler().fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1', 'x2', 'x3'])
df.plot.kde()
```

```python
data_tf = MinMaxScaler().fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1', 'x2', 'x3'])
df.plot.kde()
```

```python
data_tf = MaxAbsScaler().fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1', 'x2', 'x3'])
df.plot.kde()
```

```python
data_tf = RobustScaler().fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1', 'x2', 'x3'])
df.plot.kde()
```

```python
df = pd.DataFrame({
    'x1':np.random.chisquare(8, 1000), # 왜도 양수
    'x2':np.random.beta(8,2,1000)*40, # 왜도 음수
    'x3':np.random.normal(50, 3, 1000) # 왜도가 0
})
df.plot.kde()
```

```python
data_tf = RobustScaler().fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1', 'x2', 'x3'])
df.plot.kde()
```

```python
df = pd.DataFrame({
    'Age':[33, 44, 22, 44, 55, 22],
    'Gender':['Male', 'Female', 'Male', 'Female', 'Male', 'Male']
})
df
```

```python
df.Gender.unique()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

le = LabelEncoder() # 문자에 숫자값 할당, OneHotEncoder는 카테고리가 두개로 나뉘어질때만 사용 가능하다 -> 많을 경우에는 Dummies이용
df['gender_tf'] = le.fit_transform(df.Gender)
df
```

```python
OneHotEncoder().fit_transform(df[['gender_tf']]).toarray()
```
